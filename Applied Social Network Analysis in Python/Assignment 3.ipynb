## Part 1

Answer questions 1-4 using the network `G1`, a network of friendships at a university department. Each node corresponds to a person, and an edge indicates friendship. 

*The network has been loaded as networkx graph object `G1`.*

```python
import networkx as nx

G1 = nx.read_gml('friendships.gml')
```

### Question 1

Find the degree centrality, closeness centrality, and normalized betweeness centrality (excluding endpoints) of node 100.

*This function should return a tuple of floats `(degree_centrality, closeness_centrality, betweenness_centrality)`.*

```python
def answer_one():
        
    """
    Find the degree centrality, closeness centrality, and normalized betweeness centrality
    (excluding endpoints) of node 100.
    """
    degree_centrality = nx.degree_centrality(G1)[100]
    clsnss_centrality = nx.closeness_centrality(G1)[100]
    btwn_centrality = nx.betweenness_centrality(G1, endpoints=False, normalized=True)[100]
    
    return degree_centrality, clsnss_centrality, btwn_centrality
```

### Question 2

The voucher can be forwarded to multiple users at the same time, but the travel distance of the voucher is limited to one step, which means if the voucher travels more than one step in this network, it is no longer valid. Apply your knowledge in network centrality to select the best candidate for the voucher. 

*This function should return an integer, the name of the node.*

```python
def answer_two():
        
    """
    Suppose you are employed by an online shopping website and are tasked with selecting one user in 
    network G1 to send an online shopping voucher to. We expect that the user who receives the voucher
    will send it to their friends in the network. You want the voucher to reach as many nodes as possible.
    The voucher can be forwarded to multiple users at the same time, but the travel distance of the voucher
    is limited to one step, which means if the voucher travels more than one step in this network, it is
    no longer valid. Apply your knowledge in network centrality to select the best candidate for the voucher.
    """
    
    # compute degree centrality
    out_degree = nx.degree_centrality(G1)
    
    # find node with highest degree
    best = max(out_degree, key=out_degree.get)
    
    return best
```

## Part 2

`G2` is a directed network of political blogs, where nodes correspond to a blog and edges correspond to links between blogs. Use your knowledge of PageRank and HITS to answer Questions 5-9.

```python
G2 = nx.read_gml('blogs.gml')
```

### Question 5

Apply the Scaled Page Rank Algorithm to this network. Find the Page Rank of node 'realclearpolitics.com' with damping value 0.85.

*This function should return a float.*

```python
def answer_five():
        
    """
    Apply the Scaled Page Rank Algorithm to this network. Find the Page Rank
    of node 'realclearpolitics.com' with damping value 0.85.
    """

    # compute page rank
    SPRA = nx.pagerank(G2,alpha=0.85)
    
    # retrieve rank for 'realclearpolitics.com'
    output = SPRA['realclearpolitics.com']
    
    return output
```

### Question 6

Apply the Scaled Page Rank Algorithm to this network with damping value 0.85. Find the 5 nodes with highest Page Rank. 

*This function should return a list of the top 5 blogs in desending order of Page Rank.*

```python
def answer_six():
        
    """
    Apply the Scaled Page Rank Algorithm to this network with damping value 0.85.
    Find the 5 nodes with highest Page Rank.
    """
    import operator
    
    # compute page rank
    SPRA = nx.pagerank(G2,alpha=0.85)
    
    # sort page ranks
    pr_sorted = sorted(SPRA.items(),key=operator.itemgetter(1), reverse=True)[0:5]
    
    # convert to list
    output = [x[0] for x in pr_sorted]
    
    return output
```

### Question 7

Apply the HITS Algorithm to the network to find the hub and authority scores of node 'realclearpolitics.com'. 

*Your result should return a tuple of floats `(hub_score, authority_score)`.*

```python
def answer_seven():
        
    """
    Apply the HITS Algorithm to the network to find the hub and authority scores of node 'realclearpolitics.com'.
    """
    
    # compute hub and authority scores
    HITS = nx.hits(G2)
    
    # retrieve hub and authority score for 'realclearpolitics.com'
    hub_score = HITS[0]['realclearpolitics.com']
    authority_score = HITS[1]['realclearpolitics.com']
    
    return (hub_score, authority_score)
```

### Question 8 

Apply the HITS Algorithm to this network to find the 5 nodes with highest hub scores.

*This function should return a list of the top 5 blogs in desending order of hub scores.*

```python
def answer_eight():
        
    """
    Apply the HITS Algorithm to this network to find the 5 nodes with highest hub scores.
    """
    import operator
    
    # compute hub score
    HUB = nx.hits(G2)[0]
    
    # sort hubs
    hub_sorted = sorted(HUB.items(),key=operator.itemgetter(1), reverse=True)[0:5]
    
    # convert to list
    output = [x[0] for x in hub_sorted]
    
    return output
```
